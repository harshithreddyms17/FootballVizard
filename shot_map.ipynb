```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch
```

```python
import statsbomb as sb
comps=sb.Competitions()
events=sb.Events(event_id='3788773')
df=events.get_dataframe(event_type='shot')
df
```

```python
team1df=df[df['team']=="France"]
team2df=df[df['team']=="Portugal"]
```

```python
team1df['start_location_y'].unique
```

```python
pitch=Pitch(pitch_color='black')
fig,ax=pitch.draw(figsize=(12,8))
team1sc=pitch.scatter((team1df['start_location_x']),(team1df['start_location_y']),c='blue',s=team1df['statsbomb_xg']*500+100,edgecolor='white',ax=ax,label='France')
team2sc=pitch.scatter(120-team2df['start_location_x'],80-team2df['start_location_y'],c='green',s=team2df['statsbomb_xg']*500+100,ax=ax,edgecolor="white",label='Portugal')
ax.legend()
```
